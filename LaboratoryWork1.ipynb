```python
# переименуем столбцы в DataFrame
col_name = ['school','sex','age','address','famsize',
            'Pstatus','Medu','Fedu','Mjob','Fjob',
            'reason','guardian','traveltime','studytime','failures',
            'schoolsup','famsup','paid','activities','nursery',
            'higher','internet','romantic','famrel','freetime',
            'goout','Dalc','Walc','health','absences',
            'G1','G2','G3']
df.set_axis(col_name, axis='columns')
```

```python
# удалим строки, в которых значение в столбце 'G3'
# является NaN (пропущенным)
df.dropna(subset=['G3'], inplace=True)

# получим уникальные значения из столбца 'G3'
df.G3.unique()

# заменим все значения 'None' в столбце 'G3' на NaN
df.G3 = df.G3.replace('None', np.nan)
```
